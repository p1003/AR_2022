```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
for rank in range(9):
    arr = np.load(file=f'./results_raw/before_merge_{rank}.npy')
    plt.imshow(arr, cmap='hot', interpolation='nearest')
    plt.title(f'Result for rank {rank}')
    plt.savefig(f'./results_images/result_rank_{rank}')
```
